The `target` came out as an integer -- but inspect_ai's `multiple_choice()` solver
and `choice()` scorer expect it to be a letter (`"A"`, `"B"`, `"C"`, or `"D"`).
When the automatic mapping isn't enough, inspect_ai lets you pass a
**record-to-sample function** that receives the full raw record and returns a `Sample`
you construct yourself.

```python
def record_to_sample(record: dict) -> Sample:
    """
    Convert a raw MMLU record to an inspect_ai Sample.

    MMLU stores the correct answer as an integer index (0=A, 1=B, 2=C, 3=D).
    We convert it to the corresponding uppercase letter so it matches the
    format expected by the choice() scorer.
    """
    answer_idx = int(record["answer"])
    return Sample(
        input=record["question"],
        choices=record["choices"],
        target=ascii_uppercase[answer_idx],   # 0->'A', 1->'B', ...
        metadata=dict(subject=record.get("subject"))
    )


dataset = hf_dataset(
    path=LOCAL_PATH,
    name="all",
    split="test",
    sample_fields=record_to_sample,
    cached=True
)

sample = dataset[0]
print("target  :", sample.target, " <- letter now")
print("choices :", sample.choices)
```

## 3. Running an evaluation

Every inspect_ai evaluation is described by a `Task`, which bundles three things:

- **dataset** -- the questions
- **solver** -- the chain of steps that produces a model response;
  `multiple_choice()` formats the prompt with lettered options and parses the model's choice
- **scorer** -- the function that grades the response;
  `choice()` checks whether the selected letter matches the target

The `@task` decorator registers the function so inspect_ai can discover it by name
from the CLI or pass it directly to `eval()`.

```python
@task
def mmlu_subset(subset):
    """Minimal MMLU task for any subject subset."""
    return Task(
        dataset=subset,
        solver=[multiple_choice()],
        scorer=choice()
    )
```

Calling `eval()` runs the task and returns a **list of `EvalLog` objects** -- one per
(task, model) pair. Everything you need is in this object; there is no need to read
log files from disk.

The two most useful attributes:
- `log.results.scores` -- list of scorer results, each with a `metrics` dict
  (`"accuracy"`, `"stderr"`, etc.)
- `log.samples` -- list of `EvalSample` objects with per-question inputs, outputs, and scores

```python
logs: List[EvalLog] = eval(
    mmlu_subset(astronomy_subset),
    model=MODEL_A,
    limit=10,        # evaluate only the first 10 questions
    max_token =10
)

log = logs[0]      # one task -> one log
print("Status  :", log.status)
print("Model   :", log.eval.model)
print("Accuracy:", log.results.scores[0].metrics["accuracy"].value)
```

## Assignment 2: Implement `log_to_df`

The aggregate accuracy in `log.results` is useful for a quick check, but for the
statistical analyses ahead we need a flat table: **one row per (question, epoch)**
with a numeric `score` column.

`log.samples` is a list of `EvalSample` objects. Each one has:
- `.id` -- question identifier
- `.epoch` -- which run this belongs to (relevant when `epochs > 1`)
- `.scores` -- a dict mapping scorer name to `Score`; the `Score.value` for `choice()` is
  `"C"` (correct) or `"I"` (incorrect)
- `.metadata` -- the metadata dict you set in `record_to_sample`

Implement `log_to_df` so that it converts an `EvalLog` into a DataFrame with columns
`id`, `epoch`, `score` (1/0), and `subject`. The smoke test below will verify the shape.

```python
import pandas as pd

def log_to_df(log) -> pd.DataFrame:
    """
    """
    rows = []
    
    if not log.samples:
        return pd.DataFrame(columns=["id", "epoch", "score", "subject"])

    for sample in log.samples:
        score_val = 0
        
        # Используем современный формат доступа к оценкам через словарь .scores
        if sample.scores:
            # Берем первое значение из словаря оценок
            first_score = list(sample.scores.values())[0]
            
            # Приводим к верхнему регистру для надежного сравнения ("C", "1", "True")
            val_str = str(first_score.value).upper()
            if val_str in ["C", "1", "1.0", "TRUE"]:
                score_val = 1
        
        # Собираем строку данных
        rows.append({
            "id": sample.id,
            "epoch": getattr(sample, 'epoch', 0),
            "score": score_val,
            "subject": sample.metadata.get("subject", "unknown") if sample.metadata else "unknown"
        })
    
    return pd.DataFrame(rows)

# =================================== TESTS ===================================
df_test = log_to_df(log)

assert set(df_test.columns) >= {"id", "epoch", "score", "subject"}
assert df_test["score"].isin([0, 1]).all()
assert len(df_test) == 10

print(df_test.head())
print(f"\nAccuracy: {df_test['score'].mean():.1%}")
```

## 5. Confidence intervals


A single accuracy number carries uncertainty: the eval used a finite set of questions
sampled from a much larger space. The paper (ss2.1, ss3.1) shows how to quantify this
using the CLT standard error.


## Assignment 3: Implement `ci_accuracy_basic` and `ci_accuracy`

**`ci_accuracy_basic(scores, ci)`** -- the simple case where every question is answered
exactly once. `scores` is a plain numpy array of 0s and 1s. Use Eq. 1 from the paper.

**`ci_accuracy(df, ci)`** -- the general case that handles multiple runs per question
(`epochs > 1`). When K runs exist for a question, average their scores first, then
compute the SE across question-level averages. Pooling all K×n individual answers
would undercount variance -- answers to the same question across epochs are correlated.


```python
import numpy as np
import pandas as pd
import scipy.stats as st
from typing import Tuple

def ci_accuracy_basic(scores: np.ndarray, ci: float = 0.95) -> Tuple[float, float, float]:
    """
    CLT-based confidence interval for accuracy -- single run per question (K = 1).

    Parameters
    ----------
    scores : 1-D array of per-question binary scores (0 or 1)
    ci     : confidence level (default 0.95)

    Returns
    -------
    (lower_bound, mean_accuracy, upper_bound)
    """
    scores = np.asarray(scores)
    n = len(scores)
    mean_acc = float(np.mean(scores))
    
    # If there's 1 or fewer samples, we can't compute a meaningful variance.
    # Similarly, if variance is 0, the standard error is 0. 
    # The math naturally collapses the CI to the mean in the zero-variance case.
    if n <= 1:
        return mean_acc, mean_acc, mean_acc
        
    # Standard Error of the mean: sqrt(Var(s) / n) using sample variance (ddof=1)
    se = np.std(scores, ddof=1) / np.sqrt(n)
    
    # Z-multiplier for the requested confidence interval
    z = st.norm.ppf((1 + ci) / 2)
    
    lower_bound = mean_acc - z * se
    upper_bound = mean_acc + z * se
    
    return float(lower_bound), mean_acc, float(upper_bound)


def ci_accuracy(df: pd.DataFrame, ci: float = 0.95) -> Tuple[float, float, float]:
    """
    CLT-based confidence interval for accuracy, supporting multiple epochs (K >= 1).

    Parameters
    ----------
    df : DataFrame returned by log_to_df, with columns 'id', 'score', 'epoch'
    ci : confidence level (default 0.95)

    Returns
    -------
    (lower_bound, mean_accuracy, upper_bound)
    """
    # Average the scores for each question (id) across all runs (epochs)
    # This yields a single average score per question.
    question_means = df.groupby('id')['score'].mean().values
    
    # Pass these aggregated independent scores to the basic CI function
    return ci_accuracy_basic(question_means, ci)
```

```python
# =================================== TESTS ===================================
def _make_df(ids, scores, epochs=None):
    if epochs is None:
        epochs = [0] * len(ids)
    return pd.DataFrame({"id": ids, "score": scores, "epoch": epochs})

# ci_accuracy_basic
l, m, u = ci_accuracy_basic(np.ones(10))

assert l == 1.0 and u == 1.0, "perfect accuracy: CI should collapse to 1"

l, m, u = ci_accuracy_basic(np.zeros(10))

assert l == 0.0 and u == 0.0, "zero accuracy: CI should collapse to 0"

scores3 = np.array([1, 1, 0, 1, 0], dtype=float)
l, m, u = ci_accuracy_basic(scores3)

assert l < 0.6 < u, f"0.6 not in [{l:.3f}, {u:.3f}]"

np.random.seed(42)
s = np.random.binomial(1, 0.75, 200).astype(float)
l95, _, u95 = ci_accuracy_basic(s, 0.95)
l99, _, u99 = ci_accuracy_basic(s, 0.99)

assert (u99 - l99) > (u95 - l95), "99% CI must be wider than 95%"
assert np.isclose(l95, 0.6819421067148456, atol=10e-2)
assert np.isclose(u95, 0.8080578932851544, atol=10e-2)

# ci_accuracy (K=1 should match basic)
df3 = _make_df([1,2,3,4,5], scores3.tolist())
l_df, _, u_df = ci_accuracy(df3)
l_ar, _, u_ar = ci_accuracy_basic(scores3)

assert np.isclose(l_df, l_ar) and np.isclose(u_df, u_ar), "K=1 must match basic version"

# ci_accuracy (K=3 should give narrower CI on average)
np.random.seed(0)
rows_k1, rows_k3 = [], []
for q in range(30):
    p = np.random.uniform(0.3, 0.9)
    rows_k1.append({"id": q, "score": int(np.random.binomial(1, p)), "epoch": 0})
    for e in range(3):
        rows_k3.append({"id": q, "score": int(np.random.binomial(1, p)), "epoch": e})

l1, _, u1 = ci_accuracy(pd.DataFrame(rows_k1))
l3, _, u3 = ci_accuracy(pd.DataFrame(rows_k3))
print(f"K=1 width: {u1-l1:.3f}")
print(f"K=3 width: {u3-l3:.3f}  (narrower on average)")
print("\n✓ All tests passed!")
```

## 6. Visualising how CIs shrink

Two things make confidence intervals narrower: more questions (larger n) and more

## Assignment 4.1: Plot CI width vs number of epochs


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
from typing import List, Tuple

# ============================================================
# 1. МАТЕМАТИЧЕСКИЕ ФУНКЦИИ (CLT CI)
# ============================================================

def ci_accuracy_basic(scores: np.ndarray, ci: float = 0.95) -> Tuple[float, float, float]:
    """CLT-доверительный интервал для одной попытки на вопрос (K=1)."""
    scores = np.asarray(scores)
    n = len(scores)
    mean_acc = float(np.mean(scores))
    
    if n <= 1:
        return mean_acc, mean_acc, mean_acc
        
    # Стандартная ошибка: sqrt(Var / n). Используем ddof=1 для несмещенной оценки.
    std_dev = np.std(scores, ddof=1)
    se = std_dev / np.sqrt(n)
    
    # Z-критерий для нормального распределения
    z = st.norm.ppf((1 + ci) / 2)
    
    lower_bound = mean_acc - z * se
    upper_bound = mean_acc + z * se
    
    # Ограничиваем интервал физическими границами [0, 1]
    return max(0.0, lower_bound), mean_acc, min(1.0, upper_bound)

def ci_accuracy(df: pd.DataFrame, ci: float = 0.95) -> Tuple[float, float, float]:
    """CLT-интервал для нескольких попыток (K >= 1)."""
    # Сначала усредняем результаты внутри каждого вопроса (id), чтобы убрать корреляцию
    question_means = df.groupby('id')['score'].mean().values
    # Теперь работаем с независимыми средними значениями вопросов
    return ci_accuracy_basic(question_means, ci)

# ============================================================
# 2. ПРЕОБРАЗОВАНИЕ ДАННЫХ
# ============================================================

def log_to_df(log) -> pd.DataFrame:
    rows = []
    if not log.samples:
        return pd.DataFrame(columns=["id", "epoch", "score", "subject"])

    for sample in log.samples:
        score_val = 0
        if sample.scores:
            first_score = list(sample.scores.values())[0]
            val_str = str(first_score.value).upper()
            if val_str in ["C", "1", "1.0", "TRUE"]:
                score_val = 1
        
        rows.append({
            "id": sample.id,
            "epoch": getattr(sample, 'epoch', 0),
            "score": score_val,
            "subject": sample.metadata.get("subject", "unknown") if sample.metadata else "unknown"
        })
    return pd.DataFrame(rows)

# ============================================================
# 3. ЗАПУСК И ВИЗУАЛИЗАЦИЯ
# ============================================================

# Запуск оценки: 10 вопросов, каждый прогоняем 10 раз
logs: List[EvalLog] = eval(
    mmlu_subset(astronomy_subset),
    model=MODEL_A,
    limit=100,
    epochs=10
)

df_results = log_to_df(logs[0])

k_values   = [5, 6, 7, 8, 9, 10]
accuracies = []
ci_lowers  = []
ci_uppers  = []

for k in k_values:
    # Берем срез данных для текущего количества прогонов (K)
    df_subset = df_results[df_results['epoch'] < k]
    
    # Считаем метрики
    low, mid, high = ci_accuracy(df_subset, ci=0.95)
    
    accuracies.append(mid)
    ci_lowers.append(low)
    ci_uppers.append(high)

# Строим график
plt.figure(figsize=(8, 4))
plt.fill_between(k_values, ci_lowers, ci_uppers, alpha=0.2, color='blue', label="95% CI (CLT)")
plt.plot(k_values, accuracies, "o-", lw=2, color='blue', label="Accuracy")

plt.xlabel("Number of runs per question (K)")
plt.ylabel("Accuracy")
plt.title(f"{MODEL_A} — Accuracy & CI vs K (n=10)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Итоговая проверка
print(f"Обработано строк: {len(df_results)}")
print(f"Средняя точность на K=10: {accuracies[-1]:.1%}")
```

## 7. Comparing two models

Reporting two accuracy numbers side by side doesn't tell you whether the gap is real
or just noise. The paper (§4.2) describes a **paired test**: because both models answer
the same questions, you can compute per-question score differences and test whether
their mean differs significantly from zero. This removes question-difficulty variance
and yields a lower standard error than treating the two runs as independent samples.


## Assignment 5: Compare two models

`run_and_get_scores` and `compare_models_paired` are provided. Complete
`significance_by_paired_ttest` and use it to compare the two models on `MY_SUBSET`.

Implement `significance_by_paired_ttest` and compare MODEL_A and MODEL_B.

```python
def run_and_get_scores(model_name: str, dataset, epochs: int = 1, max_tokens: int = 100) -> np.ndarray:
    """Run eval and return mean-per-question scores, sorted by question id."""
    print(f"  Running {model_name} ...")
    run_logs = eval(mmlu_subset(dataset), model=model_name, epochs=epochs, max_tokens=max_tokens)
    df = log_to_df(run_logs[0])
    return df.groupby("id")["score"].mean().sort_index().values

def significance_by_paired_ttest(
    scores1: np.ndarray,
    scores2: np.ndarray,
    alpha: float = 0.05,
    two_tailed: bool = True,
) -> Tuple[float, float, bool]:
    """
    Paired t-test between two sets of per-question scores.
    Returns (p_value, mean_difference scores1 - scores2, is_significant).
    """
    assert len(scores1) == len(scores2), "arrays must cover the same questions"
    
    alternative = "two-sided" if two_tailed else "greater"
    
    # Используем ttest_rel для связанных выборок (paired t-test)
    _, p_value  = st.ttest_rel(scores1, scores2, alternative=alternative) # YOUR CODE HERE
    mean_diff   = np.mean(scores1) - np.mean(scores2) # YOUR CODE HERE
    
    return p_value, mean_diff, bool(p_value < alpha)

def compare_models_paired(
    model_a: str,
    model_b: str,
    dataset,
    alpha: float = 0.05,
    two_tailed: bool = True,
    epochs_a: int = 1,
    epochs_b: int = 1,
    max_tokens: int = 100,
) -> Tuple[float, float, bool]:
    """
    Evaluate both models on the same dataset and run a paired t-test.
    Returns (p_value, mean_difference A - B, is_significant).
    """
    scores_a = run_and_get_scores(model_a, dataset, epochs=epochs_a, max_tokens=max_tokens)
    scores_b = run_and_get_scores(model_b, dataset, epochs=epochs_b, max_tokens=max_tokens)
    return significance_by_paired_ttest(scores_a, scores_b, alpha, two_tailed)
```

## 9. Power analysis

Before running an expensive evaluation, it is worth asking: how many questions do we
need to detect a meaningful difference with adequate statistical power?
The paper (§5) derives the minimum detectable effect as a function of sample size n,
question-level variance ω², and within-model variance σ².

## Assignment 7: 'Estimate variance components'

Implement `estimate_variance_components` and report the MDE for `MY_SUBSET` at α = 0.05, power = 80%.

```python
def estimate_variance_components(
    logs_a: List[EvalLog],
    logs_b: List[EvalLog],
) -> dict:
    """
    Estimate omega2, sigma2_a, sigma2_b from two EvalLog objects (see ss5 of the paper).

    Both logs must cover the same set of questions. Use epochs >= 2 so that
    within-question variance can be estimated.

    Returns dict with keys 'omega2', 'sigma2_a', 'sigma2_b'.
    """

    df_a = log_to_df(logs_a[0])
    df_b = log_to_df(logs_b[0])
    mean_a = df_a.groupby("id")["score"].mean()
    mean_b = df_b.groupby("id")["score"].mean()
    var_within_a = df_a.groupby("id")["score"].var(ddof=1)
    var_within_b = df_b.groupby("id")["score"].var(ddof=1)
    common_ids = mean_a.index.intersection(mean_b.index)
    if len(common_ids) < 2:
        raise ValueError("Нужно минимум 2 общих вопроса")
    ma = mean_a.loc[common_ids].to_numpy(dtype=float)
    mb = mean_b.loc[common_ids].to_numpy(dtype=float)
    cov_matrix = np.cov(ma, mb, ddof=1)
    omega2 = float(np.var(ma, ddof=1) + np.var(mb, ddof=1) - 2.0 * cov_matrix[0, 1])
    sigma2_a = float(var_within_a.loc[common_ids].mean())
    sigma2_b = float(var_within_b.loc[common_ids].mean())
    return {
        "omega2": omega2,
        "sigma2_a": sigma2_a,
        "sigma2_b": sigma2_b,
    }


def minimum_detectable_effect(
    n: int,
    omega2: float,
    sigma2_a: float = 0.0,
    sigma2_b: float = 0.0,
    ka: int = 8,
    kb: int = 8,
    alpha: float = 0.05,
    power: float = 0.80,
) -> float:
    """MDE for a paired model comparison (Eq. 10 in the paper)."""
    z_alpha = stats.norm.ppf(1 - alpha / 2)
    z_beta  = stats.norm.ppf(power)
    return float((z_alpha + z_beta) * np.sqrt(
        (omega2 + sigma2_a / ka + sigma2_b / kb) / n
    ))
```

## Assignment 8: Implement `required_sample_size`

`minimum_detectable_effect` computes delta given n. Implement its inverse:

then use it to compute how many questions you would need to detect a 5% and a 10%
accuracy gap on `MY_SUBSET`.

```python
# --- Assignment 7 -----------------------------------------------------------
def required_sample_size(
    delta: float,
    omega2: float,
    sigma2_a: float = 0.0,
    sigma2_b: float = 0.0,
    ka: int = 1,
    kb: int = 1,
    alpha: float = 0.05,
    power: float = 0.80,
) -> int:
    """Minimum number of questions needed to detect `delta` at the given power."""
    if delta <= 0 or ka <= 0 or kb <= 0:
        raise ValueError("delta, ka, and kb must be positive")
    
    z_sum = stats.norm.ppf(1 - alpha / 2) + stats.norm.ppf(power)
    return int(np.ceil((z_sum / delta)**2 * (omega2 + sigma2_a/ka + sigma2_b/kb)))


# =================================== TESTS ===================================
n_needed = required_sample_size(delta=0.05, **params)
print(f"Questions needed to detect delta=5%: {n_needed}")

mde_check = minimum_detectable_effect(n=n_needed, **params)

assert abs(mde_check - 0.05) < 0.05, f"Round-trip failed: MDE={mde_check:.3f}"

print("Round-trip check passed!")
```

## Assignment 9: Compare a model with itself: baseline vs chain-of-thought

The `multiple_choice()` solver we have used so far prompts the model to answer
directly. inspect_ai also provides a `chain_of_thought` that asks the model

Using the paired comparison infrastructure from Section 7, evaluate the same model
twice on the same subset — once with the default solver and once with
`chain_of_thought` — and test whether the difference in accuracy is statistically
significant. Does reasoning help? Is the effect consistent across subjects?

```python
import numpy as np
import scipy.stats as st
from typing import Tuple
from inspect_ai import eval

# Импортируем нужные солверы из Inspect AI
from inspect_ai.solver import multiple_choice, chain_of_thought

def run_and_get_scores(model_name: str, dataset, plan, epochs: int = 1, max_tokens: int = 1000) -> np.ndarray:
    """Run eval with a specific solver (plan) and return mean-per-question scores."""
    # Получаем имя солвера для красивого вывода
    solver_name = plan.__name__ if hasattr(plan, "__name__") else str(plan)
    print(f"  Running {model_name} with {solver_name} (max_tokens={max_tokens})...")
    
    # Передаем plan (солвер) в eval для переопределения стандартного поведения
    run_logs = eval(
        mmlu_subset(dataset), 
        model=model_name, 
        epochs=epochs, 
        max_tokens=max_tokens,
        plan=plan
    )
    
    df = log_to_df(run_logs[0])
    return df.groupby("id")["score"].mean().sort_index().values

def significance_by_paired_ttest(
    scores1: np.ndarray,
    scores2: np.ndarray,
    alpha: float = 0.05,
    two_tailed: bool = True,
) -> Tuple[float, float, bool]:
    """
    Paired t-test between two sets of per-question scores.
    Returns (p_value, mean_difference scores1 - scores2, is_significant).
    """
    assert len(scores1) == len(scores2), "arrays must cover the same questions"
    
    alternative = "two-sided" if two_tailed else "greater"
    
    _, p_value  = st.ttest_rel(scores1, scores2, alternative=alternative)
    mean_diff   = np.mean(scores1) - np.mean(scores2)
    
    return p_value, mean_diff, bool(p_value < alpha)

def compare_solvers_paired(
    model: str,
    solver_a,
    solver_b,
    dataset,
    alpha: float = 0.05,
    two_tailed: bool = True,
    epochs: int = 5,
    max_tokens: int = 10000,
) -> Tuple[float, float, bool]:
    """
    Evaluate the SAME model on the SAME dataset using TWO DIFFERENT solvers.
    Returns (p_value, mean_difference A - B, is_significant).
    """
    # Запускаем модель с первым солвером
    scores_a = run_and_get_scores(model, dataset, plan=solver_a, epochs=epochs, max_tokens=max_tokens)
    # Запускаем ту же модель со вторым солвером
    scores_b = run_and_get_scores(model, dataset, plan=solver_b, epochs=epochs, max_tokens=max_tokens)
    
    return significance_by_paired_ttest(scores_a, scores_b, alpha, two_tailed)


# =================================== TESTS ===================================
p, d, sig = significance_by_paired_ttest(np.array([1,2,3]), np.array([1,2,3]))
assert np.isclose(d, 0.0) and not sig

p, d, sig = significance_by_paired_ttest(
    np.array([1,1,1,1,1]), np.array([0,0,0,0,0]), two_tailed=False
)
assert sig and d > 0

print("All tests passed!\n")

# ================================ ASSIGNMENT =================================
# Сравниваем direct answer и chain_of_thought для одной модели
MODEL_NAME = MODEL_A # Убедитесь, что MODEL_A и MY_SUBSET определены выше в вашем коде

p_value, diff, is_significant = compare_solvers_paired(
    model=MODEL_NAME, 
    solver_a=chain_of_thought(),  # Солвер с рассуждениями
    solver_b=multiple_choice(),   # Солвер с прямым ответом
    dataset=MY_SUBSET, 
    epochs=1,
    max_tokens=10000 # Обязательно даем больше токенов для рассуждений!
) 

print(f"P-value: {p_value:.4f}, Mean Difference (CoT - Direct): {diff:.4f}, Significant: {is_significant}")

if is_significant and diff > 0:
    print("Вывод: Chain of Thought статистически значимо улучшает результаты.")
elif is_significant and diff < 0:
    print("Вывод: Chain of Thought статистически значимо ухудшает результаты.")
else:
    print("Вывод: Статистически значимой разницы между солверами нет.")
```

## Bonus assignment: Clustered Standard Errors

Some benchmarks contain groups of related questions — for example, several questions
about the same passage in reading comprehension tasks like DROP or RACE. In such cases
the standard CLT confidence interval is anti-conservative: questions within a group are
correlated, so the effective sample size is smaller than n. Miller (2024) addresses this
with clustered standard errors (§2.2).

Using a reading comprehension benchmark of your choice, implement the clustered
confidence interval (Eq. 4 from the paper) and compare it against the naive CLT interval.
How much wider is the clustered interval? Does the difference depend on the benchmark?
Then compare two models on the same benchmark using the clustered paired standard error
(Eq. 8) — does the conclusion about which model is better change compared to the
naive paired test?

```python
import numpy as np
import pandas as pd
from scipy import stats

def get_clustered_se(scores, cluster_ids):
    # Если вы используете логи Inspect AI, это может выглядеть так:
    df_a = log_to_df(run_logs_a[0])
    df_b = log_to_df(run_logs_b[0])
    scores_a = df_a['score'].values
    scores_b = df_b['score'].values
    cluster_ids = df_a['id'].values # или другая колонка с ID пассажа
    """Реализация Eq. 4: Кластеризованная стандартная ошибка."""
    n = len(scores)
    mean_s = np.mean(scores)
    
    # Наивная ошибка (standard error of the mean)
    se_naive = np.std(scores, ddof=1) / np.sqrt(n)
    
    # Добавочный член для кластеризации
    # Группируем оценки по кластерам
    df = pd.DataFrame({'score': scores, 'cluster': cluster_ids})
    cluster_term = 0
    
    for _, group in df.groupby('cluster'):
        cluster_scores = group['score'].values
        m = len(cluster_scores)
        if m > 1:
            # Двойная сумма по i != j
            diffs = cluster_scores - mean_s
            # Сумма (si-s_mean)*(sj-s_mean) для всех i != j 
            # равна (сумма diffs)^2 - (сумма diffs^2)
            cluster_term += (np.sum(diffs)**2 - np.sum(diffs**2))
            
    se_clustered = np.sqrt(se_naive**2 + (1/n**2) * cluster_term)
    return se_naive, se_clustered

def get_confidence_interval(mean, se, alpha=0.05):
    z = stats.norm.ppf(1 - alpha/2)
    return mean - z * se, mean + z * se   
```

```python
def analyze_benchmark_results(scores_a, scores_b, cluster_ids, alpha=0.05):

    
    """
    Автоматический анализ и вывод ответов на вопросы задания.
    """
    n = len(scores_a)
    mean_a = np.mean(scores_a)
    mean_b = np.mean(scores_b)
    diff_scores = scores_a - scores_b
    mean_diff = np.mean(diff_scores)

    # 1. Считаем SE для Модели А
    se_naive_a, se_clust_a = get_clustered_se(scores_a, cluster_ids)
    
    # 2. Считаем SE для парного сравнения (Eq. 8)
    # Используем ту же функцию get_clustered_se, но передаем разность оценок
    se_naive_diff, se_clust_diff = get_clustered_se(diff_scores, cluster_ids)

    # Рассчитываем интервалы
    ci_naive_a = get_confidence_interval(mean_a, se_naive_a, alpha)
    ci_clust_a = get_confidence_interval(mean_a, se_clust_a, alpha)
    
    width_naive = ci_naive_a[1] - ci_naive_a[0]
    width_clust = ci_clust_a[1] - ci_clust_a[0]
    ratio = width_clust / width_naive

    # Тестирование гипотез (p-values)
    p_naive = 2 * (1 - stats.norm.cdf(abs(mean_diff / se_naive_diff)))
    p_clust = 2 * (1 - stats.norm.cdf(abs(mean_diff / se_clust_diff)))

    # --- ВЫВОД ОТВЕТОВ ---
    print("=== АНАЛИЗ КЛАСТЕРИЗОВАННЫХ ОШИБОК (Miller, 2024) ===")
    
    print(f"\n1. Ширина доверительного интервала:")
    print(f"   - Наивный (CLT): {width_naive:.4f}")
    print(f"   - Кластеризованный: {width_clust:.4f}")
    print(f"   - Ответ: Кластеризованный интервал шире в {ratio:.2f} раза.")
    
    if ratio > 1.1:
        print("   - Вывод: Зависимость внутри кластеров значительна. Обычный CLT слишком оптимистичен.")
    else:
        print("   - Вывод: Влияние кластеризации мало. Вопросы почти независимы.")

    print(f"\n2. Сравнение моделей (Парный тест):")
    print(f"   - Разница средних (A - B): {mean_diff:.4f}")
    print(f"   - p-value (Наивный): {p_naive:.4f}")
    print(f"   - p-value (Кластеризованный): {p_clust:.4f}")
    
    sig_naive = p_naive < alpha
    sig_clust = p_clust < alpha
    
    if sig_naive != sig_clust:
        print(f"   - Ответ: ДА, вывод изменился!")
        print(f"     При alpha={alpha} наивный тест считает разницу значимой, а кластеризованный — нет.")
    else:
        print(f"   - Ответ: НЕТ, вывод о значимости не изменился (оба теста дали {'значимый' if sig_naive else 'незначимый'} результат).")

    print("\n3. Влияние бенчмарка:")
    print("   - Если это DROP/RACE: Высокий коэффициент корреляции обусловлен тем, что несколько")
    print("     вопросов привязаны к одному тексту. Ошибка на тексте влечет ошибки во всем кластере.")
```
